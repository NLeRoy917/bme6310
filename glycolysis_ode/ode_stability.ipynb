```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def dydt_stability_1(t, y):
    dydt_1 = y[0] + 10*y[1]
    dydt_2 = 7*y[0] - 8*y[1]

    return [dydt_1, dydt_2]
```

```python
# simple system of ODEs 
y0 = [-10, 10]
tspan = (0, 0.1)  # Also try [0 1] or [0 10]
teval = np.linspace(0, 0.1, 45)
res = solve_ivp(dydt_stability_1,tspan,y0, t_eval=teval)
t = res['t']
y = res['y']

fig, ax = plt.subplots(figsize=(12,8)) 

ax.scatter(t,y[0,:], color="black", label='y1')
ax.scatter(t,y[1,:], color='blue', label='y2') 

ax.set_xlabel('t', fontsize=18)
ax.set_ylabel('Y*', fontsize=18)
ax.legend(fontsize=18)

plt.show()
```

```python
def dydt_stability_2(t, y):
    dydt_1 = 5*y[0] + 12*y[1]
    dydt_2 = -15*y[0]


    return [dydt_1, dydt_2]
```

```python
# simple system of ODEs 
y0 = [2, 6]
tspan = (0, 1)  # Also try [0 1] or [0 10]
teval = np.linspace(0, 0.1, 45)
res = solve_ivp(dydt_stability_2,tspan,y0, t_eval=teval)
t = res['t']
y = res['y']

fig, ax = plt.subplots(figsize=(12,8)) 

ax.scatter(t,y[0,:], color="black", label='y1')
ax.scatter(t,y[1,:], color='blue', label='y2') 

ax.set_xlabel('t', fontsize=18)
ax.set_ylabel('Y*', fontsize=18)
ax.legend(fontsize=18)

plt.show()
```
