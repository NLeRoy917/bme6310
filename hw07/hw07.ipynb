## Problem 1
Given that our system is an LSI system, we know that the impulse reponse is equivalent to substituting in an impulse ( $\delta[n]$ ) for $x[n]$:

$$
h[n] = \dfrac{1}{M_1 + M_2 + 1}\sum_{k=-M_1}^{M_2}\delta[n-k]
$$

Which is equivalent to:

$$
\dfrac{1}{M_1 + M_2 + 1} \Bigg(\delta[n + M_1] + \dots + \delta[n-M_2] \Bigg)
$$

Which we know, due to the definition of the impulse function will always be equal to **1** outside the interval $-M_1 \leq n \leq M_2$. Formally,

$$
h[n] = 
\begin{array}{cc}
  \Bigg \{ & 
    \begin{array}{cc}
      \dfrac{1}{M_1 + M_2 + 1} & -M_1 \leq n \leq M_2 \\
      0 & \mathrm{otherwise}
    \end{array}
\end{array}
$$

## Problem 2
We can read the input image like so:

```python
import numpy as np

im1 = plt.imread("data/cell_img.png")
im1 = np.dot(im1[..., :3], [0.299, 0.587, 0.114]) # select grayscale part of image
plt.imshow(im1, cmap="gray") # display the image
plt.show()
```

We can define both the horizontal and vertical Sobel operators like so:

```python
h_horizontal = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]]) # horizontal Sobel operator
h_vertical = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]) # vertical Sobel operator
```

We can now convolve the cell image using the `scipy.signal.convolve2d()` function.

```python
from scipy.signal import convolve2d

im1_hsobel = convolve2d(im1, h_horizontal)
im1_vsobel = convolve2d(im1, h_vertical)

fig, ax = plt.subplots(1,2)
fig.set_tight_layout({"pad": 0.5})
fig.suptitle("Cell Image convolved with Sobel operator")

ax[0].imshow(im1_hsobel, cmap="gray")
ax[0].set_title("Horizontal Sobel")

ax[1].imshow(im1_vsobel, cmap="gray")
ax[1].set_title("Vertical Sobel")
```

We can compute the final image using the pixelwise square root of the sum of the squares of the two processed images:

$$
F_{i,j} = \sum_i \sum_j \sqrt{I_{h_{i,j}} ^2 + I_{v_{i,j}} ^2}
$$

```python
final_im = np.sqrt(im1_hsobel**2 + im1_vsobel**2)

plt.imshow(final_im, cmap="gray")
```

## Problem 5
In the study og epigenomics, peak calling is a computational method used to identify areas in a genome that have been enriched with aligned reads as a consequence of performing a ChIP-sequencing or MeDIP-seq experiment. Given a genomic signal track of aligned reads, a scientist can find "peaks" of readily aligned sequences that might indicate an area of the genome that is particularly interesting to whatever experiment is being conducted (Typically this is interpreted as how _open_ chromatin are).

![Schematic of epigenomic peak calling given a signal track](figs/pcalling.png)

Often, the genomic signal can come in quite noisey and hard to analyze. This is especially true among **single-cell** experiements, where sample read sizes can tend to be quite small and thus, our data is much more prone to noise and random fluctuations. Filters using mathematical convolutions can be applied to _smooth_ outthese signal tracks and make peak calling easier for a researcher or software package. One such simple example is the **moving-average** filter:

$$
\bar{x_{i}} = \dfrac{1}{M + 1}\sum_{j=-M}^{M}x[i+j]
$$

We can achieve this result across our entire dataset by convoluting our signal track using a "window of ones". To demonstrate this I'm going to generate a simulated sinusoidal signal track that we will smooth using our moving average convolution:

$$
x[t]_{\mathrm{smooth}} = \sum_{k=0}^N x[t][\dfrac{1}{W}, \dots, \dfrac{1}{W}]
$$

Where the length of the ones vector is the size of the window, $W$.

```python

x = np.array([i for i in range(1000)])

# simulated singal function
def f_x(x: np.array) -> np.array:
    A = 20 # amplitude
    return 20*np.sin(0.01*x) + 20 + np.random.normal(-A/10, A/10, len(x))

noisy = f_x(x)

# plot the result of 
# our noist function
plt.plot(
    x, 
    noisy,
    "b-",
    linewidth=0.5
)
```

```python
from numpy import convolve

# run a convolution against our dataset usinga moving average
WINDOW = 10 # window size

# run smoothing convolution
smoothed = convolve(
    noisy, 
    np.ones(WINDOW)/WINDOW,
    "same"
)

# plot noisy
plt.plot(
    x, 
    noisy,
    "b-",
    linewidth=0.5,
    label="Noisy"
)

plt.plot(
    x, 
    smoothed,
    "r-",
    linewidth=0.5,
    label="Smoothed"
)

plt.legend()
```

We can see the smoothed signal nicely overlayed on the noisy signal which gives as a better look at the signal track and facilitates nice analysis of the data.
